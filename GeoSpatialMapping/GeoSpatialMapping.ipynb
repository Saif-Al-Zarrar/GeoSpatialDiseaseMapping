```python
import pandas as pd
import numpy as np
import sqlite3 as db
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# https://www.dataquest.io/blog/python-pandas-databases/
```

```python
tbl1= pd.read_csv("AHS_districtwise_2012-13.csv")
tbl1.head()
```

```python
if 'RURAL'  in 'State'.upper():
    print("Rural")
elif "URBAN" in 'State'.upper():
    print("Urban")
elif "TOTAL" in 'State'.upper():
    print("Total")
else:
    print("Other")
```

```python
col_names=tbl1.columns
```

```python
#Find out Which columns have urban/rural in them
col_desc=[]
for col_name in col_names:
    if 'RURAL'  in col_name.upper():
        col_desc.append("rural")
    elif "URBAN" in col_name.upper():
        col_desc.append("Urban")
    elif "TOTAL" in col_name.upper():
        col_desc.append("Total")
    else:
        col_desc.append("Other")

print(col_desc)
```

```python
col_info=pd.DataFrame({"col_names":col_names, "col_desc":col_desc}, index=None)
col_info.head()
col_info.to_csv("column_info.csv", index=False)
```
